```python
import sqlite3
import pandas as pd

# Load datasets
athletes = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/athlete_events.csv")
regions = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/noc_regions.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')


# Connect to SQLite in-memory DB
conn = sqlite3.connect(":memory:")

# Write DataFrames to SQL tables
athletes.to_sql("athletes_table", conn, index=False, if_exists="replace")
regions.to_sql("regions_table", conn, index=False, if_exists="replace")
sales.to_sql("sales_table", conn, index=False, if_exists="replace")
```

```python
#the athletes dataset
# Count the total number of medals won by each country and show the top 5.
pd.read_sql(
    """
    SELECT *
    FROM athletes_table;
    """,
    conn
)
```

```python
pd.read_sql(
    """
    SELECT Team, COUNT(Medal) as Total_Medals --Using total_medals alias to make it easy to order from highest to lowest--
    FROM athletes_table
    WHERE Medal != 'None' --ensuring columns with 'none' are not counted--
    GROUP BY Team --to print output in countries--
    ORDER BY Total_Medals DESC --recording from highst to lowest number--
    LIMIT 5;
    """,
    conn
)
#
```

```python
#Calculate the average age of athletes who won a Gold medal.
pd.read_sql(
    """
    SELECT AVG(Age) as Average_Age
    FROM athletes_table
    WHERE Medal = 'Gold'; --ensuring only those with gold are counted--
    """,
    conn
)
```

```python
#How many distinct events are there in each sport?
pd.read_sql(
    """
    SELECT DISTINCT Event, Sport
    FROM athletes_table
    GROUP BY Sport; --satisfying the each sport parameter--
    """,
    conn
)
```

```python
#Show all athletes from the United States (NOC = 'USA').
pd.read_sql(
    """
    SELECT DISTINCT Name, NOC --disctinct keyword so that an athlete appears only once--
    FROM athletes_table
    WHERE NOC = 'USA';
    """, conn
)
```

```python
#Count how many medals were awarded each year.
pd.read_sql(
    """
    SELECT COUNT(Medal), Year --showing the number of medals plus year of award--
    FROM athletes_table
    WHERE Medal != 'None' --excludes all entries with none--
    GROUP BY Year;
    """, conn
)
```

```python
#Find all athlete records where height or weight is missing.
pd.read_sql(
    """SELECT *
    FROM athletes_table
    WHERE Height IS NULL OR Weight IS NULL; --checking both height or weight in one line--
    """, conn
)
```

```python
#Replace the missing height with the average athlete height.
pd.read_sql(
    """
    SELECT ID, Name, Age, Weight, Team, Height,
    COALESCE(Height, (SELECT AVG(Height) FROM athletes_table)) AS New_Height --replace the height with average height value where missing--
    FROM athletes_table;
    """, conn
)
```

```python
#The sales dataset: where necessary, use the key word CAST to remove the $ sign from the item_price
df_prices = pd.read_sql(
    """
    SELECT *,
    CAST(REPLACE(item_price, '$', '') AS REAL) AS Prices --replacing $ with nothing--
    FROM sales_table;
    """, conn
)
df_prices.head()
```

```python
#Return total sales value per item.
pd.read_sql(
    """
    SELECT item_name, SUM(quantity * CAST(REPLACE(item_price, '$', '') AS REAL)) AS total_sales_value
    FROM sales_table
    GROUP BY item_name;
    """, conn
)
```

```python
#Show the top 5 records with the highest item_price
pd.read_sql(
    """
    SELECT CAST(REPLACE(item_price, '$', '') AS REAL) AS prices
    FROM sales_table
    ORDER BY prices DESC
    LIMIT 5;
    """, conn
)
```

```python
#How many unique customer orders are there? (Assume each order_id is a customer.)
pd.read_sql(
    """
    SELECT COUNT(DISTINCT order_id) AS unique_orders
    FROM sales_table;    
    """, conn
)
```
